```python
from moviepy import VideoFileClip, AudioFileClip, TextClip, CompositeVideoClip, ColorClip
from moviepy.audio.fx import AudioFadeIn, AudioFadeOut
from pathlib import Path

```

```python
def split_audio_file(audio_path, output_path, start_time, end_time):
    audio_clip = AudioFileClip(audio_path)
    audio_clip.subclipped(start_time, end_time).write_audiofile(output_path)
    audio_clip.close()

#test
org_audio_path = "/Users/sangdo/Downloads/audio/original/disco_haingoai_pbszVbZ6yq4.mp3"
test_output_path = "/Users/sangdo/Downloads/audio/disco/BoiTinLoiThe_2.mp3"
start_time = 7*60 + 18
# split_audio_file(org_audio_path, test_output_path, start_time, start_time + 40)   #very fast
```

```python
def merge_audio_video(audio_path, video_path, output_path):
    video_clip = VideoFileClip(video_path)
    audio_clip = AudioFileClip(audio_path)
    audio_clip = audio_clip.with_effects([    #avoid sudden sound from the beginning and end
        AudioFadeIn(1),
        AudioFadeOut(1)
    ])

    final_clip = video_clip.with_audio(audio_clip)  #replace audio from video
    #add waterark
    # Background box
    bg_height = 80
    bg = (
        ColorClip(size=(int(video_clip.w * 0.8), bg_height), color=(0, 0, 0))
        .with_opacity(0.6)
        .with_duration(video_clip.duration)
        .with_position(("center", video_clip.h - bg_height - 40))
    )

    text = (
        TextClip(
            text="https://www.facebook.com/nhac.chua.lanh",
            font_size=30,
            color="white",
            method="caption",
            size=(int(video_clip.w * 0.8), bg_height)
        )
        .with_duration(video_clip.duration)
        .with_position(("center", video_clip.h - bg_height - 45))
    )

    final_clip = CompositeVideoClip([final_clip, bg, text])

    final_clip.write_videofile(output_path, codec="libx264", audio_codec="aac")
    
    video_clip.close()
    audio_clip.close()
    final_clip.close()
    
#test files
test_audio_path = "/Users/sangdo/Downloads/audio/disco/BoiTinLoiThe_2.mp3"
test_video_path = "/Users/sangdo/Downloads/video/pexels/8745380-hd_1080_1920_25fps.mp4"
test_output_path = "/Users/sangdo/Downloads/video/output/output.mp4"
merge_audio_video(test_audio_path, test_video_path, test_output_path)   #30s video takes approx. 1m15s to create
```
