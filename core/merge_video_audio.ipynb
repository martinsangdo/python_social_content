```python
from moviepy import VideoFileClip, AudioFileClip
```

```python
#test files
test_audio_path = "/Users/sangdo/Downloads/audio/disco/BoiTinLoiThe.m4a"
test_video_path = "/Users/sangdo/Downloads/video/city/3320517-uhd_2160_3840_25fps.mp4"
test_output_path = "/Users/sangdo/Downloads/video/output/output.mp4"
```

```python
def merge_audio_video(audio_path, video_path, output_path):
    video_clip = VideoFileClip(video_path)
    audio_clip = AudioFileClip(audio_path)
    
    final_clip = video_clip.with_audio(audio_clip)
    final_clip.write_videofile(output_path, codec="libx264", audio_codec="aac")
    
    video_clip.close()
    audio_clip.close()
    
#test
merge_audio_video(test_audio_path, test_video_path, test_output_path)
```
