<h2>Split video into small files</h2>

```python
original_video = PARENT_PATH + 'video/youtube/_original_long/nweEhc0M-Jc.mp4' #long video

def split_video(start_sec, output_video_path):
    end_sec = start_sec + 30    #30 seconds
    Path(output_video_path).parent.mkdir(parents=True, exist_ok=True)

    with VideoFileClip(original_video) as video:
        end_sec = min(end_sec, video.duration)

        clip = video.subclipped(start_sec, end_sec)

        clip.write_videofile(
            output_video_path,
            codec="libx264",
            audio_codec="aac"
        )
#test
start_sec = 0 * 60 + 50
# split_video(start_sec, PARENT_PATH + 'video/output/nweEhc0M-Jc_2.mp4')
```

<h2>Split audio into small files</h2>

```python
def split_audio_file(audio_path, output_path, start_time, end_time):
    audio_clip = AudioFileClip(audio_path)
    audio_clip.subclipped(start_time, end_time).write_audiofile(output_path)
    audio_clip.close()
```

```python
original_long_audio_filename = 'geVG6BU8RkU'   #mp3
original_long_audio = PARENT_PATH + 'audio/_original_long/'+original_long_audio_filename + '.mp3' #long audio
#test
start_minute = 0
start_second = 0
start_time = start_minute*60 + start_second
new_filename = original_long_audio_filename + '_'+str(start_minute)+'_'+str(start_second)+'.mp3'
test_output_path = PARENT_PATH + "audio/disco/"+new_filename
split_audio_file(original_long_audio, test_output_path, start_time, start_time + 30)   #very fast
```

<h2>Merge audio and video</h2>

```python
audio_30secs = PARENT_PATH + 'audio/disco/ĐờiChiaLyGiữaLốiGaVắng.mp3'

video_30secs = PARENT_PATH + 'video/youtube/5uES_EJiB44_1.mov'
final_video_output_path = PARENT_PATH + "video/output/5uES_EJiB44_1.mp4"
```

```python
def merge_audio_video(audio_path, video_path, output_path):
    video_clip = VideoFileClip(video_path)
    audio_clip = AudioFileClip(audio_path)
    audio_clip = audio_clip.with_effects([    #avoid sudden sound from the beginning and end
        AudioFadeIn(1),
        AudioFadeOut(1)
    ])

    final_clip = video_clip.with_audio(audio_clip)  #replace audio from video
    #add waterark
    # Background box
    bg_height = 100
    bg = (
        ColorClip(size=(int(video_clip.w * 0.8), bg_height), color=(0, 0, 0))
        .with_opacity(0.6)
        .with_duration(video_clip.duration)
        .with_position(("center", video_clip.h - bg_height - 40))
    )

    text = (
        TextClip(
            text="Người Ta Sẽ Nói Ôm Chua Cay Vì Em ...",
            font_size=60,
            color="white",
            method="caption",
            size=(int(video_clip.w * 0.8), bg_height)
        )
        .with_duration(video_clip.duration)
        .with_position(("center", video_clip.h - bg_height - 60))
    )

    # final_clip = CompositeVideoClip([final_clip, bg, text])

    final_clip.write_videofile(output_path, codec="libx264", audio_codec="aac")
    
    video_clip.close()
    audio_clip.close()
    final_clip.close()
    
#test files
merge_audio_video(audio_30secs, video_30secs, final_video_output_path)   #30s video takes approx. 1m15s to create
```
