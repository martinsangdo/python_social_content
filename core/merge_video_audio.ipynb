```python
from moviepy import VideoFileClip, AudioFileClip, TextClip, CompositeVideoClip, ColorClip
```

```python
def merge_audio_video(audio_path, video_path, output_path):
    video_clip = VideoFileClip(video_path)
    audio_clip = AudioFileClip(audio_path)
    
    final_clip = video_clip.with_audio(audio_clip)
    #add waterark
    # Background box
    bg_height = 80
    bg = (
        ColorClip(size=(int(video_clip.w * 0.8), bg_height), color=(0, 0, 0))
        .with_opacity(0.6)
        .with_duration(video_clip.duration)
        .with_position(("center", video_clip.h - bg_height - 40))
    )

    text = (
        TextClip(
            text="https://www.facebook.com/nhac.chua.lanh",
            font_size=48,
            color="white",
            method="caption",
            size=(int(video_clip.w * 0.8), bg_height)
        )
        .with_duration(video_clip.duration)
        .with_position(("center", video_clip.h - bg_height - 45))
    )

    final_clip = CompositeVideoClip([final_clip, bg, text])

    final_clip.write_videofile(output_path, codec="libx264", audio_codec="aac")
    
    video_clip.close()
    audio_clip.close()
    final_clip.close()
    
#test
# merge_audio_video(test_audio_path, test_video_path, test_output_path)   #30s video takes approx. 1m15s to create
```
