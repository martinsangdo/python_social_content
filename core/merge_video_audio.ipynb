```python
def merge_img_2_video(img_path, video_path, output_path):
    #the video should have middle bg as #eee, width = 1040, height = 1350, x: 20, y: 495
    video = VideoFileClip(video_path)

    image = (
        ImageClip(img_path)
        .with_duration(video.duration)
        .with_position(("center", 565))
    )

    final = CompositeVideoClip([video, image])

    final.write_videofile(
        output_path,
        codec="libx264",
        audio_codec="aac"
    )

    video.close()
    final.close()

#test
test_img_path = "/Users/sangdo/Downloads/math_games_video/img/addition_1.png"
test_video_path = "/Users/sangdo/Downloads/math_games_video/template/video_template_2.mp4"
test_output_path = "/Users/sangdo/Downloads/math_games_video/output/addition_1.mp4"
# merge_img_2_video(test_img_path, test_video_path, test_output_path) #approx. 30s to create
```
