```python
def merge_img_2_video(img_path, video_path, output_path):
    #the video should have middle bg as #eee, width = 1040, height = 1350, x: 20, y: 495
    video = VideoFileClip(video_path)

    image = (
        ImageClip(img_path)
        .with_duration(video.duration)
        .with_position(("center", 565))
    )

    final = CompositeVideoClip([video, image])

    final.write_videofile(
        output_path,
        codec="libx264",
        audio_codec="aac"
    )

    video.close()
    final.close()

#test
test_output_path = "/Users/sangdo/Downloads/math_games_video/output/20260304.mp4"
merge_img_2_video(selected_img, template_video, test_output_path) #approx. 30s to create
```

```python
def capture_video_2_img(video_path, output_path, time_sec):
    Path(output_path).parent.mkdir(parents=True, exist_ok=True)

    with VideoFileClip(video_path) as video:
        frame = video.get_frame(time_sec)  # numpy array
        video.save_frame(output_path, t=time_sec)

#test
img_capture_path = "/Users/sangdo/Downloads/math_games_video/output/capture_img/20260304.png"
capture_video_2_img(test_output_path, img_capture_path, 10)
```
