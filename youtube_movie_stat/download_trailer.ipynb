```python
import pymongo
import os
import subprocess

from dotenv import load_dotenv
load_dotenv(override=True) 

db_client = pymongo.MongoClient(os.environ['REMOTE_MONGO_DB'])
db = db_client['db_movie']   
```

<h2>Download clip 20s from youtube trailer</h2>

```python
PROJECT_FOLDER = '/Users/sangdo/Downloads/movie-top-list/'
TRAILER_FOLDER = PROJECT_FOLDER + 'trailers/'
```

```python
def download_1_trailer(trailer_id, folder_name):
    url = "https://www.youtube.com/watch?v=" + trailer_id
    output = TRAILER_FOLDER + folder_name + "/" + trailer_id + ".mp4"
    #download and cut 20s from the video
    cmd = [
        "yt-dlp",
        "--download-sections", "*00:00:40-00:01:00",
        "--force-keyframes-at-cuts",
        "-f", "bv*[width=640]",   #best video only, no audio "bv*+ba/b"
        "--merge-output-format", "mp4",
        "-o", output,
        url
    ]
    subprocess.run(cmd, check=True)
    print('Finish downloading: ' + trailer_id)

def download_trailers(actor_name, folder_name):
    tb_paycheck = db['tb_paycheck']
    data = tb_paycheck.find({'actor': actor_name})

    for item in data:
        download_1_trailer(item['trailer'], folder_name)

#test
# download_trailers('Tom Cruise', 'tom_cruise')
# download_1_trailer('MPr0OyV9_c0', 'tom_cruise')
```
